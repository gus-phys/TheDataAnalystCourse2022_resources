# Preprocessing
---

## ID

```python
df['ID'].nunique()
```

```python
df.drop('ID', axis=1, inplace=True)
```

Since `ID` do not contribute to estimations and can't used as index either, this columns can be droped

## Reason for Absense

```python
df['Reason for Absence'].nunique()
```

```python
print(sorted(df['Reason for Absence'].unique()))
```

The reason number 20 is missing or nobody claimed it.

**.get_dummies**

```python
reason_columns = pd.get_dummies(df['Reason for Absence'], drop_first=True)
reason_columns
```

## Group the reasons for absence

```python
reason_type_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_type_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_type_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_type_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
df.drop('Reason for Absence', axis=1, inplace=True)

df = pd.concat([df, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis=1)

df
```

**Rename columns**

```python
df.rename({
    0: 'Reason_1',
    1: 'Reason_2',
    2: 'Reason_3',
    3: 'Reason_4'
}, axis=1, inplace=True)
```

**Reorder columns**

```python
df.columns
```

```python
column_names_reorderd = [
  'Reason_1', 
  'Reason_2', 
  'Reason_3',
  'Reason_4', 
  'Date', 
  'Transportation Expense', 
  'Distance to Work', 
  'Age',
  'Daily Work Load Average', 
  'Body Mass Index', 
  'Education', 
  'Children',
  'Pets',
  'Absenteeism Time in Hours'
]
```

```python
df = df[column_names_reorderd]
```

**Checkpoint!**

```python
df_reason_mod = df.copy()
```

```python
df_reason_mod
```

## Date

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format='%d/%m/%Y')
```

```python
type(df_reason_mod['Date'][0])
```

**Extracting the month value**

```python
list_months = []
for i in range(len(df_reason_mod)):
  list_months.append(df_reason_mod['Date'][i].month)

# list_months
```

```python
df_reason_mod.insert(5, 'Month Value', list_months)

df_reason_mod.head()
```

**Extracting day of the week**

```python
list_weekday = []
for i in range(len(df_reason_mod)):
  list_weekday.append(df_reason_mod['Date'][i].weekday())

# list_weekday
```

```python
df_reason_mod.insert(6, 'Day of the Week', list_weekday)

# df_reason_mod.head()
```

**Checkpoint!**

```python
df_reason_date_mod = df_reason_mod.copy()

df_reason_date_mod
```

## Education

```python
df_reason_date_mod['Education'].value_counts()
```

```python
df_reason_date_mod['Educatino'] = df_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```
