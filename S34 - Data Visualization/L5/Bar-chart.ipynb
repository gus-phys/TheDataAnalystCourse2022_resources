### Libraries

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

### Data

```python
df = pd.read_csv(data_dir+"/bar_chart_homework_data.csv")

df
```

# Bar chart
---

```python
plt.figure(figsize=(9,6))

plt.bar(x = df["Cities"],
        height = df["Frequency"],
        color="midnightblue")

plt.title("Frequency per cities", fontsize=16, fontweight="bold")
plt.ylabel("Frequency", fontsize=13)
plt.xticks(rotation=45, fontsize=13)
plt.xticks(fontsize=13)

plt.savefig(current_dir+"/Frequency.png", bbox_inches='tight')
plt.show()
```
