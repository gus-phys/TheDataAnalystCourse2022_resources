### Libraries

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

### Data

```python
df = pd.read_csv(data_dir+"/area_chart_homework_data.csv", usecols=[0,1,2,3], )

df
```

# Area chart
---

**Add a column for months**

```python
list_dates = pd.to_datetime(df['Date'])

list_months = []
for date in list_dates:
  list_months.append(date.strftime("%B"))

df.insert(1, 'Month', list_months)
```

```python
df
```

## Stack Plot

```python
# colors = ["#011638", "#7e2987", "ef2026"]
sns.set_style("white")

plt.figure(figsize=(12,6))

plt.stackplot(df['Month'],
              df['International'],
              df['Domestic'],
              df['Services'],
              # color=colors,
              edgecolor='none')

plt.title("Stacked Area Chart of 2020", fontsize=16, weight="bold")
plt.xlabel("Month", fontsize=13)
plt.ylabel("Value", fontsize=13)
plt.xticks(rotation=45)
plt.legend(labels=df.columns[2:], loc='upper left')

sns.despine()

plt.savefig(current_dir+"/Stacked_plot.png", bbox_inches='tight')
plt.show()
```
